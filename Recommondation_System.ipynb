# Product recommendation Systems


## **What are Recommendation Systems?**

A recommendation system, also known as a recommender system, is a type of software or algorithmic approach that provides personalized suggestions to users, aiming to help them discover items of interest. These systems are widely used in various domains, including e-commerce, entertainment, social media, content platforms, and more. The primary goal of recommendation systems is to enhance user experience by presenting relevant and appealing options based on the user's preferences, behavior, or characteristics.

Recommendation systems operate by analyzing available data, such as user interactions, preferences, historical behaviors, and item attributes. They then generate recommendations that aim to predict what items a user might like or find valuable. There are several techniques and approaches used in building recommendation systems, including:

1. Collaborative Filtering:
Collaborative filtering is based on the idea that users who have agreed in the past tend to agree again in the future. It relies on user-item interaction data to identify similarities between users or items. There are two main types of collaborative filtering:

User-Based Collaborative Filtering: Recommends items to a user based on the preferences of users with similar tastes.
Item-Based Collaborative Filtering: Recommends items to a user based on the preferences of items that the user has interacted with before.
<br>

2. Content-Based Filtering:
Content-based filtering recommends items based on the attributes of items and the preferences of the user. It relies on item characteristics or features and matches them with the user's profile or historical interactions.

3. Hybrid Methods:
Hybrid methods combine multiple techniques to improve recommendation quality. For example, a hybrid system might merge collaborative filtering and content-based approaches to leverage their respective strengths.

4. Matrix Factorization:
Matrix factorization techniques decompose the user-item interaction matrix into latent factors that capture underlying patterns in the data. This is particularly useful for dealing with sparsity in the interaction matrix.

5. Deep Learning:
Deep learning-based recommendation systems use neural networks to learn intricate patterns and relationships from user behavior and item attributes. They can capture complex interactions and are suitable for large-scale data.

6. Context-Aware Recommendation:
Context-aware recommendation considers additional contextual information, such as time, location, or user context, to generate more relevant and timely recommendations.

7. Reinforcement Learning:
Reinforcement learning-based recommendation treats recommendation as a sequential decision-making problem, where the system learns to maximize user satisfaction over time.

8. Evaluation Metrics:
To assess the effectiveness of a recommendation system, various metrics are used, including precision, recall, F1-score, Mean Average Precision (MAP), and user engagement metrics.

Effective recommendation systems provide numerous benefits, including increased user engagement, improved customer satisfaction, enhanced conversion rates, and better content discovery. However, building an effective recommendation system requires careful consideration of data quality, algorithm choice, user privacy, and ongoing model updates to ensure accurate and relevant recommendations.

```python
#calculate the density of rating matrix
givenratings = np.count_nonzero(ratings_matrix)
print('given num of ratings = ', givenratings)
possibleratings = ratings_matrix.shape[0] * ratings_matrix.shape[1]
print('possible num of ratings = ', possibleratings)
density = (givenratings/possibleratings)
density *= 100
print ('density: {:4.2f}%'.format(density))
```

# Building Popularity Recommonder Model

training the dataset

```python
#recommondation score
train_data_grouped = train_data.groupby('item_id').agg({'user_id': 'count'}).reset_index()
train_data_grouped.rename(columns = {'user_id': 'score'},inplace=True)
train_data_grouped.head(10)
```

```python
# sorting products based on recommondation score
train_data_sort = train_data_grouped.sort_values(['score', 'item_id'], ascending = [0,1]) 
# generating rank based on recommondation score
train_data_sort['rank'] = train_data_sort['score'].rank(ascending=0, method='first') 

#printing top 5 recommondations
popular_recommendations = train_data_sort.head(5) 
popular_recommendations
```

```python
#using popularity based recommender model to make predictions
def recommend(user_id):     
    user_recommendations = popular_recommendations 
          
    #Add user_id column for which the recommendations are being generated 
    user_recommendations['user_id'] = user_id 
      
    #Bring user_id column to the front 
    cols = user_recommendations.columns.tolist() 
    cols = cols[-1:] + cols[:-1] 
    user_recommendations = user_recommendations[cols] 
          
    return user_recommendations
```

*Since, it is a Popularity recommender model, so, all the five users are given the same recommendations. Here, we predict the products based on the popularity. It is not personalized to particular user. It is a non-personalized recommender system.

#  Building Collaborative Filtering Recommendor Model

```python
colab_user = pd.concat([train_data, test_data]).reset_index()
colab_user.head()
```

```python
# Matrix with row per 'user' and column per 'item' 
pivot_df = colab_user.pivot(index = 'user_id', columns ='item_id', values = 'rating').fillna(0)
pivot_df.head()
```

```python
print('Shape of the pivot table: ', pivot_df.shape)
```

* since the above matrix is sparse matrix we will use SVD

**Singular Value Decomposition (SVD)** is a matrix factorization technique that has various applications in recommender systems. SVD decomposes a matrix into three matrices, capturing latent factors that can represent underlying patterns in the data. In the context of recommender systems, SVD can be applied to user-item interaction matrices to uncover hidden features and generate personalized recommendations

SVD is a foundational technique in *collaborative filtering*. Collaborative filtering relies on user-item interaction data to identify similar users or items and make recommendations. By applying SVD to the interaction matrix, it decomposes the matrix into user and item latent factor matrices. These latent factors capture hidden patterns and relationships in the data. The predicted values obtained from the decomposed matrices can be used to suggest items to users who haven't interacted with them before.

```python
from scipy.sparse.linalg import svds
from scipy.sparse import csc_matrix
```

```python
# singular value Decomposition
# U, sigma, Vt = svds(pivot_df, k = 10)

# Assuming pivot_df is your sparse matrix

# Perform singular value decomposition
# U, sigma, Vt = svds(pivot_df, k=10)

# # Make sure to sort singular values in descending order
# sigma = np.diag(sigma[::-1])

# # Now you can reconstruct the original matrix if needed
# original_matrix = np.dot(np.dot(U, sigma), Vt)



# Convert the pivot table to a sparse matrix
pivot_sparse = csc_matrix(pivot_df)

# Perform singular value decomposition
k = 10  # Choose the number of singular values to compute
U, sigma, Vt = svds(pivot_sparse, k=k)

# Make sure to sort singular values in descending order
sigma = np.diag(sigma[::-1])



```

```python
print('Left singular matrix: \n\n',U)

print('\n Shape of U: ', U.shape)
```

```python
print('Sigma: \n\n',sigma)

print('\nShape of sigma: ', sigma.shape)
```

```python
Sigma = np.diag(sigma)
print('Diagonal matrix: \n\n',sigma)

print('Shape of the sigma: ', Sigma.shape)
```

```python
print('Right singular matrix: \n\n',Vt)

print('Shape of the Vt: ', Vt.shape)
```

```python
# predicted ratings
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) 

# Convert predicted ratings to dataframe

preds_df = pd.DataFrame(all_user_predicted_ratings, columns = pivot_df.columns)
preds_df.head()
```

```python
# recommend items with highest predicted ratings
def recommend_items(user_id, pivot_df, preds_df, num_recommendations):
    sorted_user_ratings = pivot_df.iloc[user_id].sort_values(ascending=False)
    sorted_user_predictions = preds_df.iloc[user_id].sort_values(ascending=False)
    temp = pd.concat([sorted_user_ratings, sorted_user_predictions], axis=1)
    temp.index.name = 'Recommended Items'
    temp.columns = ['user_ratings', 'user_predictions']
    temp = temp.loc[temp.user_ratings == 0]   
    temp = temp.sort_values('user_predictions', ascending=False)
    print('\nBelow are the recommended items for user(user_id = {}):\n'.format(user_id))
    print(temp.head(num_recommendations))
```
